### Load the ner-multi model

```python
from flair.data import Sentence
from flair.models import SequenceTagger

# load tagger
tagger = SequenceTagger.load("flair/ner-multi")
```

## DutchPolicyDocs

```python
file_path = '../../../data/DutchPolicyDocs/DutchPolicyDocs.json' 

import loading_functions

toponym_data = loading_functions.prepare_data(file_path, filtered=False, split=False)
```

### Predictions for DutchPolicyDocs

```python
import flair_predictions

processed_predictions_all = flair_predictions.make_predictions(toponym_data, tagger)
```

### Results DutchPolicyDocs

```python
import evaluate

# All toponyms
fps, fns = evaluate.evaluate(toponym_data, processed_predictions_all)
```

```python
fps
```

```python
fns
```
