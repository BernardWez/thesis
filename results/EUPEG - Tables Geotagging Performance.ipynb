```python
import pandas as pd
```

```python
df = pd.read_excel('geoparsing-results-EUPEG.xlsx')
df.head()
```

### F score rank

```python
fscore_df = df.pivot(index='geoparser', columns='dataset', values='f-score')
rank_df = fscore_df.rank(method='min', ascending=False).astype(int)
rank_df['total'] = rank_df.sum(axis=1)
rank_df.sort_values('total')
```

```python
fscore_var = fscore_df.var(axis=1).sort_values(ascending=False)
fscore_mean = fscore_df.mean(axis=1).sort_values(ascending=False)

print('variance')
print(fscore_var)
print()
print('mean')
print(fscore_mean)
```

```python
precision_df = df.pivot(index='geoparser', columns='dataset', values='precision')
precision_var = precision_df.var(axis=1).sort_values(ascending=False)
precision_mean = precision_df.mean(axis=1).sort_values(ascending=False)

print('variance')
print(precision_var)
print()
print('mean')
print(precision_mean)
```

```python
recall_df = df.pivot(index='geoparser', columns='dataset', values='recall')
recall_var = recall_df.var(axis=1).sort_values(ascending=False)
recall_mean = recall_df.mean(axis=1).sort_values(ascending=False)

print('variance')
print(recall_var)
print()
print('mean')
print(recall_mean)
```
